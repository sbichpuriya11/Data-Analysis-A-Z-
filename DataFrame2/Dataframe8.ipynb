```python
foods = pd.read_csv('../foods.csv')
foods.head()
```

```python
foods['Item'] = foods['Item'].astype('category')
foods['Gender'] = foods['Gender'].astype('category')
foods['City'] = foods['City'].astype('category')
foods['Frequency'] = foods['Frequency'].astype('category')
```

```python
foods.info()
```

```python
foods.sort_values('First Name',inplace=True)
```

```python
foods
```

```python
foods[foods['First Name'].duplicated(keep='first')]
```

```python
foods['First Name'].duplicated(keep='last')
```

```python
foods['First Name'].duplicated(keep='last')
foods[foods['First Name'].duplicated(keep='last')]
```

```python
~foods['First Name'].duplicated(keep=False)
```

```python
foods[~foods['First Name'].duplicated(keep=False)]
```
