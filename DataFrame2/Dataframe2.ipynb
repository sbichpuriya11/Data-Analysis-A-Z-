```python
df = pd.read_csv('../employees.csv')
df.head(4)
```

```python
df['Gender'] = df['Gender'].astype('category')
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['Team'] = df['Team'].astype('category')
df['Senior Management'] = df['Senior Management'].astype('bool')
df.head()
```

```python
df[df['Gender'] == 'Male']
```

```python
df[df['Team'] =='Finance']
```

```python
marketing = df["Team"]=='Marketing'
```

```python
df[marketing]
```

```python
salary = df['Salary']>100000
df[salary]
```

```python
df[df['Senior Management']]
```

```python
domain = df['Team'] !='Product'
df[domain]
```

```python
df[df['Bonus %']<1.5]
```

```python
date = df['Start Date']<='1-20-1984'
df[date]
```

```python
df['Last Login Time'] = pd.to_datetime(df['Last Login Time'])
```

```python
time  = df['Last Login Time'] == '1:40 AM'
df[time]
```
