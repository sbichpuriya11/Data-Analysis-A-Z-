# Modifying Plot Aesthetics with matplotlib Templates

```python
import matplotlib.pyplot as plt
import pandas as pd
from  pandas_datareader import data
```

```python
bb = data.DataReader(name='BB',data_source='yahoo',start='2004-01-01',end='2020-12-31')
```

```python
bb.head()
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
```

```python
bb.plot(y='Close')
```

```python
plt.style.use('dark_background')
bb.plot(y='Close')
```

```python
plt.style.use('ggplot')
bb.plot(y='Close')
```
