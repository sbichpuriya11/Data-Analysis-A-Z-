# Create Range of dates with the pd.date_range() Method, Part 2

```python
import pandas as pd
```

```python
pd.date_range(start = '2012-09-09',periods=25,freq='D')
```

```python
pd.date_range(start = '2012-09-09',periods=50,freq='D')
```
