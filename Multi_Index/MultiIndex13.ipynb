```python
sales = pd.read_csv('../salesmen.csv',parse_dates=['Date'])
sales.head()
```

```python
len(sales)
```

```python
sales['Salesman'].value_counts()
```

```python
sales['Salesman'] = sales['Salesman'].astype('category')
sales.head()
```

```python
sales.pivot(index='Date',columns = 'Salesman',values="Revenue")
```

```python
s = pd.read_csv('../worldstats.csv')
s['country'] = s['country'].astype('category')
s['year']  = s['year'].astype('category')
s.head()
```

```python
len(s)
```

```python
s.pivot(index='year',columns='country',values=['Population','GDP'])
```

```python
len(s.pivot(index='year',columns='country',values=['Population','GDP']))
```
