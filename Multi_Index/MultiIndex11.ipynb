# The .unstack() Method, Part 2

```python
import pandas as pd
```

```python
world = pd.read_csv('../worldstats.csv',index_col=['country','year'])
world.head()
```

```python
s = world.stack()
s.head()
```

```python
s.unstack(2)
s.unstack(0)
```

```python
s.unstack(1)
```

```python
s.unstack(-1)
```

```python
s.unstack(-1)
s.unstack(2)
```

```python
s.unstack(-2)
s.unstack(1)
```

```python
s.unstack(-3)
s.unstack(0)
```

```python
s.unstack('year')
s.unstack('country')
```
