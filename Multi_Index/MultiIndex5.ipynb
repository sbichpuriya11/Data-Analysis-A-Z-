# The sort_index Method on a MultiIndex Dataframe

```python
import pandas as pd
```

```python
bigmac = pd.read_csv('../bigmac.csv',parse_dates=['Date'],index_col=['Date','Country'])
bigmac.head()
```

```python
bigmac.sort_index()
bigmac.sort_index(ascending=False)
```

```python
bigmac.sort_index(ascending=[False,True]) #this will sort the date in asc while country in dsc
```

```python
bigmac.sort_index(level=0,ascending=True)
bigmac.sort_index(level=['Date'],ascending=True)
```

```python
bigmac.sort_index(level=1)
bigmac.sort_index(level='Country', ascending=False)
```

```python
bigmac.sort_index(level=['Date','Country'],ascending=[False,True])
bigmac.sort_index(level=['Date','Country'],ascending=[True,False]) 
bigmac.sort_index(level=['Date','Country'],ascending=[True,False],inplace=True) 
```

```python
bigmac.sort_index(level=[1,0],ascending=[True,False],inplace=True)
bigmac.head()
```
