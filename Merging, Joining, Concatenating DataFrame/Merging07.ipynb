```python
week1 = pd.read_csv('../Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('../Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('../Restaurant - Customers.csv')
food = pd.read_csv('../Restaurant - Foods.csv')
```

```python
merged = week1.merge(right=week2,how='left',on='Customer ID',indicator=True)
```

```python
pd.set_option('display.max_row',None)
week1.merge(right=food,how='left',on='Food ID')
```

```python
pd.set_option('display.max_row',None)
week1.merge(right=food,how='left',on='Food ID',sort=True) #will be sorted on the basis 'how' values
```
