```python
import pandas as pd
```

```python
week1 = pd.read_csv('../Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('../Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('../Restaurant - Customers.csv',index_col='ID')
food = pd.read_csv('../Restaurant - Foods.csv',index_col='Food ID')
```

```python
customers.head()
```

```python
food.head()
```

```python
week1.head()
```

```python
sales = week1.merge(right=customers,how='left',left_on='Customer ID',right_index=True)
```

```python
sales.head()
```

```python
sales = sales.merge(food,how='left',left_on='Food ID',right_index=True)
```

```python
week1.merge(week2,how='left',left_index=True,right_index=True)
```

```python
week1.merge(week2,how='left',left_index=True,right_index=True,suffixes=[' - Week 1',' - Week 2'])
```
