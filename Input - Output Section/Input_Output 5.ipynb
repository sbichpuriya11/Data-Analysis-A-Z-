# Export excel File

```python
import pandas as pd
```

```python
baby_names = pd.read_csv('https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv')
```

```python
baby_names.head()
```

```python
males = baby_names[baby_names['Gender']=='MALE']
males
```

```python
females = baby_names[baby_names['Gender']== 'FEMALE']
females
```

```python
excel_file = pd.ExcelWriter('Baby_Names.xlsx')
```

```python
males.to_excel(excel_file,sheet_name='Boys',index=False)
```

```python
females.to_excel(excel_file,sheet_name='Girls',index=False,columns=['Year of Birth','Gender','Ethnicity'])
```

```python
excel_file.save()
```
