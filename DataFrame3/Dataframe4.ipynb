```python
bond = pd.read_csv('../jamesbond.csv')
bond.head()
```

```python
bond.iloc[0]
bond.iloc[10]
```

```python
bond.iloc[[15,4]]
# bond.iloc[100]
```

```python
bond.iloc[4:8:2]
```

```python
bond.set_index('Film',inplace=True)
bond.sort_index(inplace=True)
bond.head()
```

```python
bond.iloc[17]
```

```python
bond.iloc[1:9]
```

```python
bond.iloc[1:10:2]
```

```python
bond.iloc[[12,3,15,10]]
# bond.iloc[[12,3,15,10,90]]
```
