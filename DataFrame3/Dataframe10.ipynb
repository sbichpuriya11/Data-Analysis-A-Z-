```python
bond = pd.read_csv('../jamesbond.csv',index_col='Film')
bond.sort_index(inplace=True)
bond.head()
```

```python
bond.sample(n=3)
#either use n or frac both means same (fraction is the percent of data we need eg. 45% of rows, to use frac just write percent in percent/100 form)
bond.sample(frac=.34)
```

```python
# values for axis 'axis0/ rows', 'axis1/columns' ,'index'
bond.sample(n=4,axis= 'columns')
bond.sample(n=4,axis= 'rows')
bond.sample(n=2,axis='index')
```
