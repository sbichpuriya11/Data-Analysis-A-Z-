```python
bond = pd.read_csv('../jamesbond.csv',index_col='Film')
bond.sort_index(inplace=True)
bond.head()
```

```python
bond.sort_values('Box Office',ascending=False).head(4)
```

```python
bond.nlargest(n=3,columns='Box Office')
```

```python
bond.nsmallest(n=2,columns='Budget')
```

```python
bond.nlargest(n=3,columns='Budget')
```

```python
bond.nsmallest(4,'Bond Actor Salary')
```

```python
bond['Box Office'].nlargest(n=8)
```

```python
bond['Year'].nsmallest(2)
```
