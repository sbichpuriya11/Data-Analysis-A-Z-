```python
bond = pd.read_csv('../jamesbond.csv',index_col='Film')
bond.sort_index(inplace=True)
bond.head()
```

```python
def add_million(n):
    return str(n)+' M'
bond['Box Office'] = bond['Box Office'].apply(add_million)
```

```python
bond['Budget'] = bond['Budget'].apply(add_million)
```

```python
def add_million(n):
    if(n):
        return str(n)+' K'
```

```python
bond = pd.read_csv('../jamesbond.csv',index_col='Film')
bond.sort_index(inplace=True)
bond.head()
```

```python
columns=['Box Office','Bond Actor Salary','Budget']
for col in columns:
    bond[col] = bond[col].apply(add_million)
    
```
