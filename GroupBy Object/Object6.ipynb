```python
fortune = pd.read_csv('../fortune1000.csv' ,index_col='Rank')
sectors = fortune.groupby(by=['Sector'])
fortune
```

```python
sectors['Employees'].mean()
```

```python
sectors.agg({
    'Revenue':'sum',
    'Profits':'mean',
    'Employees':'sum'
})
```

```python
sectors.agg(['size','sum','mean','max'])
```

```python
sectors.agg({
    'Revenue':['sum','mean'],
    'Profits':'mean',
    'Employees':'sum'
})
```
